1. Genera un DataFrame con los datos de los cuatro archivos

```python
#Declaro las bibliotecas que necesitaré para la elaboración del ejercicio
import pandas as pd
import numpy as np
import io
#librería que ayuda a la manipulación de fechas y hora
import datetime as dt
```

```python
from google.colab import files
```

```python
emi_2016=files.upload()
emi_2017=files.upload()
emi_2018=files.upload()
emi_2019=files.upload()
```

```python
e_2016=pd.read_csv(io.BytesIO(emi_2016["emisiones-2016.csv"]),sep=";", header=0)
e_2017=pd.read_csv(io.BytesIO(emi_2017["emisiones-2017.csv"]),sep=";", header=0)
e_2018=pd.read_csv(io.BytesIO(emi_2018["emisiones-2018.csv"]),sep=";", header=0)
e_2019=pd.read_csv(io.BytesIO(emi_2019["emisiones-2019.csv"]),sep=";", header=0)

Emisiones = pd.concat([e_2016, e_2017, e_2018, e_2019])
Emisiones
```

2. Filtrar las columnas del DataFrame, quedarse con las columnas "Estación, Magnitud, Año, Mes" y los correspondientes a los dias D01, D02, etc.

```python
col = ['ESTACION', 'MAGNITUD', 'ANO', 'MES']#Columnas que quiero
col.extend([i for i in Emisiones if i.startswith('D')])#Con ayuda del DataFrame, recorro su contenido con un contador "i" para así, dejar las columnas que necesito. i.startswith('D') es para dejar todos los nombres que inicien con "D" 
Emisiones = Emisiones[col]
Emisiones
```

3. Reestructurar el DataFrame para que los valores contaminantes de las columnas de los días aparezcan en una única columna

```python
Emisiones = Emisiones.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name='VALOR')#La función melt sirve para dar formato a un DataFrame a una o más columnas
print(Emisiones)
```

4. Añadir una nueva columna con la fecha a partir de la concatenación del año, el mes y día

```python
Emisiones["DIA"]= Emisiones.DIA.str.strip('D')#la función .strip elimina cualquier caracter inicial o final
Emisiones['FECHA'] = Emisiones.ANO.apply(str) + '/' + Emisiones.MES.apply(str) + '/' + Emisiones.DIA.apply(str) #Concatenamos las columnas

Emisiones['FECHA'] = pd.to_datetime(Emisiones.FECHA, format='%Y/%m/%d', infer_datetime_format=True, errors='coerce')#coerce lo que hace es que si hay valores fuera de los límites colocará NaT en su lugar, no usamos "ignore"(visto en clase) porque solo nos dará la entrada original
print(Emisiones)
```

5. Eliminar las filas con fechas no válidas y ordenar el DataFrame por estaciones contaminantes y fecha.

```python
Emisiones = Emisiones.drop(Emisiones[np.isnat(Emisiones.FECHA)].index)#Eliminamos los valores no válidos
Emisiones.sort_values(['ESTACION', 'MAGNITUD', 'FECHA'])#sort_values hace la clasificación más rápida
```

6. Mostar las estaciones y contaminantes disponibles en el DataFrame

```python
print("Estaciones:", Emisiones.ESTACION.unique(), "\nContaminantes:", Emisiones.MAGNITUD.unique() )#unique sirve para ver los elementos que están dentro de un arreglo
```

7. Crear una función que reciba una estación, un contaminante y rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estación y rango de fechas dado.

```python
def Fun(estacion, contaminante, fecha1, fecha2):
    return Emisiones[(Emisiones.ESTACION == estacion) & (Emisiones.MAGNITUD == contaminante) & (Emisiones.FECHA >= fecha1) & (Emisiones.FECHA <= fecha2)].sort_values('FECHA').VALOR
```

```python
Fun(56, 8, dt.datetime.strptime('2019/08/01', '%Y/%m/%d'), dt.datetime.strptime('2019/12/31', '%Y/%m/%d'))#Se manda a llamar la función, .strptime crea un formato de fecha de acuerdo con una cadena previamente proporcionada
```

10. Crear una función que reciba una estación y un contaminante y devuelva un resumen descriptivo de las emisiones del contaminante indicado en la estación indicada

```python
def resumendes(estacion, contaminante):
    return Emisiones[(Emisiones.ESTACION == estacion) & (Emisiones.MAGNITUD == contaminante)].VALOR.describe()#La función describe hace que se muestre la descripción de los datos ingresados por el usuario
```

```python
print('Resumen Descriptivo: \n', resumendes(4, 6),'\n', sep='')
```

11. Crear una función que devuelva las emisiones medias mensuales de un contaminante y un año dados para todas las estaciones

```python
def EMM(contaminante, años):
    return Emisiones[(Emisiones.ANO == años) & (Emisiones.MAGNITUD == contaminante)].groupby(['ESTACION', 'MES']).VALOR.mean().unstack('MES')#unstack sirve para que se vaya moviendo entre los meses
```

```python
EMM(6, 2016)
```
