```python
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42
    }

sse = []
for k in range(1, 11):
  kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
  kmeans.fit(scaled_features)
  sse.append(kmeans.inertia_)

plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(range(1, 11), sse, curve="convex", direction="decreasing")
kl.elbow
```

```python
# A list holds the silhouette coefficients for each k
silhouette_coefficients = []
   
# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 11):
  kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
  kmeans.fit(scaled_features)
  score = silhouette_score(scaled_features, kmeans.labels_)
  silhouette_coefficients.append(score)

plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Generamos algunos datos aleatorios para demostraciÃ³n
X = df.drop(columns=['type'])

# Instanciamos el objeto PCA con el nÃºmero de componentes deseados
pca = PCA(n_components=2)

# Ajustamos el modelo PCA a nuestros datos y transformamos los datos a las nuevas dimensiones
X_pca = pca.fit_transform(X)

# Imprimimos los resultados
print("Varianza explicada por cada componente:", pca.explained_variance_ratio_)
print("Varianza total explicada:", sum(pca.explained_variance_ratio_))
print("Dimensiones originales de los datos:", X.shape)
print("Dimensiones transformadas de los datos:", X_pca.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import pandas as pd

data = df.copy()

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(data.drop('type', axis=1), data['type'], test_size=0.3)

# Crear un clasificador utilizando regresiÃ³n logÃ­stica
clf = LogisticRegression()

# Entrenar el clasificador utilizando el conjunto de entrenamiento
clf.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = clf.predict(X_test)

# Generar el classification report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Definir los datos de entrada (X) y las etiquetas 
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Dividir los datos en conjunto de entrenamiento y de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo SVM
svm_model = SVC(kernel='linear')

# Entrenar el modelo con los datos de entrenamiento
svm_model.fit(X_train, y_train)

# Hacer predicciones con los datos de prueba
y_pred = svm_model.predict(X_test)

# Mostrar el reporte de clasificaciÃ³n
print(classification_report(y_test, y_pred))
```
